```python
# Name a more iconic trio... I'll wait…
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# get data for gold coin 

# dollar 
gold_price = 1500
colletralized_percentage = 80
rebuys = 100
range_rebuys = range(1,rebuys+1)
nan_array = np.full(rebuys, np.nan)
```

```python
df = pd.DataFrame.from_dict(dict([('# gold coins', range_rebuys), ('own money invested', nan_array), ('loaned money', nan_array), ('price of assets', nan_array), ('gold price', nan_array)]), dtype=np.double)

df = df.astype({'# gold coins': int})

df.set_index('# gold coins', inplace=True, drop=True)
df
```

```python

current_assets = 0
current_money_invested = 0
current_loaned_money = 0
gold_price = 1500

for index, row in df.iterrows():
    current_money_invested += (gold_price * (100-colletralized_percentage)) / 100
    current_loaned_money += (gold_price * colletralized_percentage) / 100
    current_assets += gold_price

    row['own money invested'] = current_money_invested
    row['loaned money'] = current_loaned_money
    row['price of assets'] = current_assets
    

df.plot()
```
